```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('../Data/Sales_January_2019.csv')
df1['month']='jan'

df2=pd.read_csv('../Data/Sales_February_2019.csv')
df2['month']='feb'

df3=pd.read_csv('../Data/Sales_March_2019.csv')
df3['month']='mar'

df4=pd.read_csv('../Data/Sales_April_2019.csv')
df4['month']='apr'

df5=pd.read_csv('../Data/Sales_May_2019.csv')
df5['month']='may'

df6=pd.read_csv('../Data/Sales_June_2019.csv')
df6['month']='june'

df7=pd.read_csv('../Data/Sales_July_2019.csv')
df7['month']='july'

df8=pd.read_csv('../Data/Sales_August_2019.csv')
df8['month']='oug'

df9=pd.read_csv('../Data/Sales_September_2019.csv')
df9['month']='sep'

df10=pd.read_csv('../Data/Sales_October_2019.csv')
df10['month']='oct'

df11=pd.read_csv('../Data/Sales_November_2019.csv')
df11['month']='nov'

df12=pd.read_csv('../Data/Sales_December_2019.csv')
df12['month']='dec'
```

```python
list = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12]
```

```python
frame = pd.concat(list, ignore_index = True)
frame
```

```python
frame.info()
```

```python
frame.isnull().sum()
```

```python
frame.nunique()
```

```python
frame['Product'].unique()
```

```python
frame.columns = [col.lower() for col in frame]
frame.columns
```

```python
frame.rename(columns={
    'order id'         : 'order_id',
    'quantity ordered' : 'quantity_ordered',
    'price each'       : 'price_each',
    'order date'       : 'order_date',
    'purchase address' : 'purchase_address',
}, inplace = True
            )
frame
```

```python
(frame.isnull().sum().sum())/len(frame)*100
```

```python
frame = frame.dropna()
frame.isnull().sum()
```

```python
frame['quantity_ordered'].unique()
```

```python
frame['price_each'].unique()
```

```python
df = frame[frame['quantity_ordered'] != 'Quantity Ordered']
```

```python
df['quantity_ordered'].unique()
```

```python
df = frame[frame['price_each'] != 'Price Each']
```

```python
df['price_each'].unique()
```

```python
df = df.astype({
    'quantity_ordered' : int,
    'price_each'       : float
}
                    )
```

```python
df.info()
```

```python
df['sales'] = df['quantity_ordered'] * df['price_each']
df
```

```python
month_sales = df.groupby('month')['sales'].agg(sales_sum = 'sum').sort_values(by='sales_sum', ascending= False).reset_index()
month_sales
```

```python
fig, ax = plt.subplots(figsize=(9,6))
ax = sns.barplot(x ='sales_sum',
                  y='month',
                data = month_sales,
                palette='viridis')

plt.title('Month wise Sale')

for container in ax.containers:
    ax.bar_label(container, fmt='%.2f')
    
plt.show
```

```python
def get_city(address):
    return address.split(',')[1].strip(' ')

def get_state(address):
    return address.split(',')[2].split(' ')[1]

df['city'] = df['purchase_address'].apply(lambda x: get_city(x) + '(' + get_state(x) + ')')

df.head()
```

```python
city_sales = df.groupby('city')['sales'].agg(sales_sum = 'sum').sort_values(by='sales_sum', ascending= False).reset_index()
city_sales
```

```python
fig, ax = plt.subplots(figsize=(9,6))
ax = sns.barplot(x ='sales_sum',
                  y='city',
                data = city_sales,
                palette='viridis')

plt.title('City wise Sale')

for container in ax.containers:
    ax.bar_label(container, fmt='%.2f')
    
plt.show
```

```python
product_sales = df.groupby('product')['quantity_ordered'].agg(quantity_ordered = 'sum').sort_values(by="quantity_ordered", ascending=False).reset_index()
product_sales
```

```python
fig, ax = plt.subplots(figsize=(9, 6))

ax = sns.barplot(x = "quantity_ordered",
            y = "product",
            data = product_sales,
                palette='viridis')

plt.title("Prouct and Quantity Ordered")

for container in ax.containers:
    ax.bar_label(container, fmt='%.2f')
plt.show()
```

```python
time_sales = df
list_time = []
for i in time_sales['order_date']:
    list_time.append(i.split(" ")[1])
    
time_sales['time'] = list_time
time_sales.head()
```

```python
time_sales = df.groupby('time')['sales'].agg(time_sales = 'sum').sort_values(by="time_sales", ascending=False).reset_index()
time_sales.head()
```

```python
fig, ax = plt.subplots(figsize=(9, 6))

ax = sns.barplot(x = "time_sales",
            y = "time",
            data = time_sales.head(),
                palette='viridis')

plt.title("Prouct Sales By Time")

for container in ax.containers:
    ax.bar_label(container, fmt='%.2f')
plt.show()
```
